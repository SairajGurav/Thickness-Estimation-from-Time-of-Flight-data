```python
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Concatenate, Flatten, Dense

def inception_net_regression(input_shape):       ## Inception-like model

    inputs = Input(input_shape)

    # Convolutional block 1
    conv1_1 = Conv1D(8, 1, activation='relu', padding='same')(inputs)
    conv1_2 = Conv1D(16, 1, activation='relu', padding='same')(inputs)
    conv1_3 = Conv1D(32, 1, activation='relu', padding='same')(inputs)
    concat1 = Concatenate(axis=-1)([conv1_1, conv1_2, conv1_3])
    #pool1 = MaxPooling1D(pool_size=2)(concat1)

    # Convolutional block 2
    conv2_1 = Conv1D(64, 1, activation='relu', padding='same')(concat1)
    conv2_2 = Conv1D(128, 3, activation='relu', padding='same')(concat1)
    conv2_3 = Conv1D(256, 5, activation='relu', padding='same')(concat1)
    concat2 = Concatenate(axis=-1)([conv2_1, conv2_2, conv2_3])
    #pool2 = MaxPooling1D(pool_size=2)(concat2)

    # Convolutional block 3
    conv3_1 = Conv1D(64, 1, activation='relu', padding='same')(concat2)
    conv3_2 = Conv1D(128, 1, activation='relu', padding='same')(conv3_1)
    conv3_3 = Conv1D(256, 1, activation='relu', padding='same')(conv3_2)
    #concat3 = Concatenate(axis=-1)([conv3_1, conv3_2, conv3_3])
    # #pool3 = MaxPooling1D(pool_size=2)(concat3)

    # Flatten
    flatten = Flatten()(conv3_3)

    # Dense layers
    dense1 = Dense(256, activation='relu')(flatten)
    dense2 = Dense(128, activation='relu')(dense1)

    # Output layer
    output = Dense(1, activation='linear')(dense2)

    # Create model
    model = Model(inputs=inputs, outputs=output)

    return model
```

```python
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Conv1D, MaxPooling1D, Concatenate, Flatten, Dense

def inception_net_classification(input_shape):       ## Inception-like model

    inputs = Input(input_shape)

    # Convolutional block 1
    conv1_1 = Conv1D(8, 1, activation='relu', padding='same')(inputs)
    conv1_2 = Conv1D(16, 1, activation='relu', padding='same')(inputs)
    conv1_3 = Conv1D(32, 1, activation='relu', padding='same')(inputs)
    concat1 = Concatenate(axis=-1)([conv1_1, conv1_2, conv1_3])
    #pool1 = MaxPooling1D(pool_size=2)(concat1)

    # Convolutional block 2
    conv2_1 = Conv1D(64, 1, activation='relu', padding='same')(concat1)
    conv2_2 = Conv1D(128, 3, activation='relu', padding='same')(concat1)
    conv2_3 = Conv1D(256, 5, activation='relu', padding='same')(concat1)
    concat2 = Concatenate(axis=-1)([conv2_1, conv2_2, conv2_3])
    #pool2 = MaxPooling1D(pool_size=2)(concat2)

    # Convolutional block 3
    conv3_1 = Conv1D(64, 1, activation='relu', padding='same')(concat2)
    conv3_2 = Conv1D(128, 1, activation='relu', padding='same')(conv3_1)
    conv3_3 = Conv1D(256, 1, activation='relu', padding='same')(conv3_2)
    #concat3 = Concatenate(axis=-1)([conv3_1, conv3_2, conv3_3])
    # #pool3 = MaxPooling1D(pool_size=2)(concat3)

    # Flatten
    flatten = Flatten()(conv3_3)

    # Dense layers
    dense1 = Dense(256, activation='relu')(flatten)
    dense2 = Dense(128, activation='relu')(dense1)

    # Output layer
    output = Dense(14, activation='softmax')(dense2)

    # Create model
    model = Model(inputs=inputs, outputs=output)

    return model
```

```python
classification_model = inception_net_classification(input_shape = (17, 1))
classification_weights_path = '/content/drive/MyDrive/data for raml/Euler_type/models/classification_model (1).h5'  # Specify the path to the weights file
classification_model.load_weights(classification_weights_path)
```

```python
from tensorflow.keras.optimizers import Adam
regression_model = inception_net_regression(input_shape=(17, 1))

learning_rate = 0.001
optimizer = Adam(learning_rate=learning_rate)
#regression_model.compile(loss='mean_absolute_error', optimizer= optimizer)
```

```python
regression_model.summary()
```

```python
classification_weights = classification_model.get_weights()[:-3]

regression_weights = regression_model.get_weights()

regression_weights[:len(classification_weights)] = classification_weights
regression_model.set_weights(regression_weights)
```

```python
regression_model.compile(loss='mean_absolute_error', optimizer= optimizer)
```

```python
regression_model.fit(X_train, y_train, epochs=25, batch_size=32, validation_data=(X_val, y_val))
```

```python
# Evaluate the model on the test data
loss = regression_model.evaluate(X_test, y_test)
print("Test loss:", loss)
```
